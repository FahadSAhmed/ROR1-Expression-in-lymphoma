```python
# prompt: import /content/drive/MyDrive/Colab Notebooks/FCCC/New Analysis DLBCL MCL/DLBCL/combined.csv

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FCCC/New Analysis DLBCL MCL/DLBCL/combined.csv')
print(df.head())
```

```python
num_samples = df.shape[0]
print(num_samples)
```

```python
# Sort dataframe by ROR1 in descending order
df_sorted = df.sort_values(by="ROR1", ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(range(len(df_sorted)), df_sorted["ROR1"], label="ROR1", color='red')

plt.xlabel("Samples (n=2238)")
plt.ylabel("TPM Expression Level")
plt.title("TPM Expression of ROR1 in Diffuse Large B-cell Lymphoma (NCBI-GEO datasets: GSE117556, GSE181063)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set y-axis lower bound to 160 and upper bound slightly above the max value
plt.ylim(160, df_sorted["ROR1"].max() * 1.05)

plt.tight_layout()
plt.show()
```
