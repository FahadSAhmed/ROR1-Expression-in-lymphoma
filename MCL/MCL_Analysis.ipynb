```python
# If running in Google Colab or a local environment without bioinfokit, uncomment:
# !pip install bioinfokit

import pandas as pd
import numpy as np
from bioinfokit.analys import norm

#------------------------------------------------------------------------------
# 1. Read each file into a DataFrame
#    We assume tab-delimited with 'GeneSymbol' as a column for gene IDs.
#------------------------------------------------------------------------------
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FCCC/MCL_ROR1/MCL_microarray_gene_symbols_1.txt',
                  sep='\t', index_col='GeneSymbol')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FCCC/MCL_ROR1/MCL_microarray_gene_symbols_3.txt',
                  sep='\t', index_col='GeneSymbol')
df3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FCCC/MCL_ROR1/MCL_microarray_gene_symbols_5.txt',
                  sep='\t', index_col='GeneSymbol')
df4 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FCCC/MCL_ROR1/MCL_microarray_gene_symbols_7.txt',
                  sep='\t', index_col='GeneSymbol')
df5 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FCCC/MCL_ROR1/MCL_microarray_gene_symbols_8.txt',
                  sep='\t', index_col='GeneSymbol')

#------------------------------------------------------------------------------
# 2. Remove any duplicated gene symbols in each DataFrame
#    (keeping the first occurrence, or you could average duplicates instead).
#------------------------------------------------------------------------------
df1 = df1[~df1.index.duplicated(keep='first')]
df2 = df2[~df2.index.duplicated(keep='first')]
df3 = df3[~df3.index.duplicated(keep='first')]
df4 = df4[~df4.index.duplicated(keep='first')]
df5 = df5[~df5.index.duplicated(keep='first')]

#------------------------------------------------------------------------------
# 3. Take the intersection of genes across all DataFrames so that
#    we only keep genes present in all five studies.
#------------------------------------------------------------------------------
common_genes = set(df1.index) & set(df2.index) & set(df3.index) & set(df4.index) & set(df5.index)
common_genes = list(common_genes)  # Convert set to list to avoid indexing error

#------------------------------------------------------------------------------
# 4. Subset each DataFrame to these common genes (same row order).
#------------------------------------------------------------------------------
df1 = df1.loc[common_genes]
df2 = df2.loc[common_genes]
df3 = df3.loc[common_genes]
df4 = df4.loc[common_genes]
df5 = df5.loc[common_genes]

#------------------------------------------------------------------------------
# 5. Concatenate all data side by side: rows = genes, columns = samples.
#------------------------------------------------------------------------------
df_combined = pd.concat([df1, df2, df3, df4, df5], axis=1)

#------------------------------------------------------------------------------
# 6. Create a batch list that matches the columns in df_combined.
#    For example, each DataFrame is one batch. If df1 has 10 columns, you get
#    10 labels of "Batch1", etc.
#------------------------------------------------------------------------------
batch1 = ['Batch1'] * df1.shape[1]
batch2 = ['Batch2'] * df2.shape[1]
batch3 = ['Batch3'] * df3.shape[1]
batch4 = ['Batch4'] * df4.shape[1]
batch5 = ['Batch5'] * df5.shape[1]

batch_list = batch1 + batch2 + batch3 + batch4 + batch5

#------------------------------------------------------------------------------
# 7. Transpose so each row is a sample and each column is a gene
#    (required by ComBat in bioinfokit).
#------------------------------------------------------------------------------
df_combined_t = df_combined.T  # shape: samples x genes

#------------------------------------------------------------------------------
# 8. Run ComBat for batch correction.
#    By default, norm.combat() expects log-transformed data (e.g., log2).
#    If you have additional covariates, pass them with 'covar=' as well.
#------------------------------------------------------------------------------
combat_output = norm.combat(
    data=df_combined_t,    # DataFrame, each row = sample, columns = genes
    batch=batch_list,      # list of batch labels
    covar=None             # or define a covariate list if needed
)

#------------------------------------------------------------------------------
# 9. Transpose back so rows are genes, columns are samples.
#------------------------------------------------------------------------------
df_corrected = combat_output.T  # shape: genes x samples

#------------------------------------------------------------------------------
# 10. Export or continue with downstream analysis.
#    The data are now batch-corrected.
#------------------------------------------------------------------------------
df_corrected.to_csv('/content/batch_corrected_expression.txt', sep='\t')

print("Batch correction complete!")
print("df_corrected shape:", df_corrected.shape)
```
